```python
import os
import pandas as pd
```

```python
df_train = pd.read_csv('F:/Test_Environment/Project/YOLO/Getting_Data/train-annotations-bbox.csv')
```

```python
df_test =pd.read_csv('F:/Test_Environment/Project/YOLO/Getting_Data/test-annotations-bbox.csv')
```

```python
# select the id based on the image required - use class-descriptions-boxable.csv
image_id = '/m/0k4j' #----id for cars
num_of_train_samples = 800
num_of_test_samples = 200
```

```python
# location of the train data csv file
train_filename = 'F:/Test_Environment/Project/YOLO/Getting_Data/train-annotations-bbox.csv'

# create the file image_list_file
image_list_file_path = os.path.join('.', 'image_list_file')

# list to hold values
image_id_list = []


print(train_filename)
with open(train_filename, 'r') as f:
    line = f.readline()
    while len(line) != 0:
        id, _, label_name, _, xmin, xmax, ymin, ymax= line.split(',')[:8]
        if label_name == image_id and id not in image_id_list:
            image_id_list.append(id)
            # to write to image list file
            with open(image_list_file_path, 'a') as fw:
                fw.write('{}/{}\n'.format('train', id))
        # read next line
        line = f.readline()
        # check num of samples and break
        if len(image_id_list) == num_of_train_samples:
            break
    f.close()
```

```python
# Run the code in terminal to download the images 
# $IMAGE_LIST_FILE - image list file name - add path of file
# $DOWNLOAD_FOLDER - name of folder in which images should be downloaded
# CODE:  python downloader.py $IMAGE_LIST_FILE(path) --download_folder=$DOWNLOAD_FOLDER(path) --num_processes=5
# python downloader.py F:/Test_Environment/Project/YOLO/Getting_Data/image_list_file --download_folder=F:/Test_Environment/Project/YOLO/Data/images/train --num_processes=5
```

```python
df_train = df_train[df_train['LabelName']==image_id]
for img in image_id_list:
    df_train_img = df_train[df_train['ImageID']==img]
    for i in range(df_train_img.shape[0]):
        line_df = df_train_img.iloc[i]
        xmin, xmax, ymin, ymax = line_df['XMin'], line_df['XMax'], line_df['YMin'], line_df['YMax']
        annotation_path = 'F:/Test_Environment/Project/YOLO/Data/labels/train/' + img + '.txt'
        with open(annotation_path, 'a') as annf:
                xmin, xmax, ymin, ymax = [float(j) for j in [xmin, xmax, ymin, ymax]]
                xc = (xmin + xmax) / 2
                yc = (ymin + ymax) / 2
                w = xmax - xmin
                h = ymax - ymin
                annf.write('0 {} {} {} {}\n'.format(xc, yc, w, h))
                annf.close()
        
```
