### Testing on Images and Annotating It.

```python
# Define directories and file paths
IMAGES_DIR = os.path.join(ROOT_DIR, 'unseen_img')
model_path = os.path.join(ROOT_DIR, 'runs', 'detect', 'train2', 'weights', 'last.pt')

# Load YOLO model
model = YOLO(model_path)

# Define threshold for object detection
threshold = 0.2
```

```python
def annotate_image(image_name):
  # getting the image and displaying it
  image_path = os.path.join(IMAGES_DIR, image_name)
  image = mpimg.imread(image_path)
  plt.imshow(image)
  plt.axis('off')
  plt.show()

  output_image_path = os.path.join(IMAGES_DIR, image_name[:-4]+'_out.jpg')

  # Read image
  image = cv2.imread(image_path)

  # Perform object detection
  results = model(image)
  # getting the boundaries of the box
  image_results = results[0]

  annotation_values = image_results.boxes.data.tolist()
  for i in annotation_values:
    x1, y1, x2, y2, score, class_id = i
    # creating the annotation
    if score > threshold:
      cv2.rectangle(image, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
      cv2.putText(image, image_results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)
    cv2.imwrite(output_image_path, image)

  #displaying the image after annotation
  image = mpimg.imread(output_image_path)

  plt.imshow(image)
  plt.axis('off')
  plt.show()
```
